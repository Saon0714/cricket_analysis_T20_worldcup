```python
# Importing all the required modules
import pandas as pd
import json
```

**Match Summary**

```python
# reading the json file
with open("/content/t20_wc_match_results.json") as f:
  data = json.load(f)
data
```

```python
# Converting the json file to pandas dataframe
df_match= pd.DataFrame(data[0]["matchSummary"])
df_match.head()
```

```python
# Shape of the dataframe
df_match.shape
```

```python
# Renaming the scorecard column to match_id
df_match.rename({"scorecard":"match_id"},axis=1,inplace=True)
df_match.head()
```

```python
df_match.info()
```

```python
# Converting matchDate column to datetime datatype
df_match["matchDate"]=pd.to_datetime(df_match["matchDate"])
df_match.head()
```

```python
# Creating a new column based on a specific date
df_match["Stage"]=df_match["matchDate"].apply(lambda x:"Qualifier" if x<pd.to_datetime('2022-10-22') else "Super 12")
```

```python
df_match.head()
```

```python
# Exporting the Dataframe as a csv file
df_match.to_csv("Match_summary_csv.csv",index=False)
```

**Batting Summary**

```python
# Creating a pandas dataframe from the batting summary json file
with open("/content/t20_wc_batting_summary.json") as f:
  data=json.load(f)
  all_records=[]
  for rec in data:
    all_records.extend(rec["battingSummary"])
df_batting=pd.DataFrame(all_records)
df_batting.head()
```

```python
# Removing the unnecessary strings from batsmanName column
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x : x.replace("(c)",""))
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x : x.replace("",""))

df_batting.head(10)
```

```python
# Creating a column where the batsman is out or not based on the dismissal column
df_batting["Out"]=df_batting["dismissal"].apply(lambda x:1 if len(x)>0 else 0)
df_batting.head(10)
```

```python
# Dropping the dismissal column as we don't that now
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head(10)
```

```python
# Renaming teamInnings column to team
df_batting.rename(columns={"teamInnings":"team"},inplace=True)
df_batting.head()
```

```python
# Now to connect the df_match dataframe to df_batting we need to create a dictionary 
match_id_dict={}
for index,rec in df_match.iterrows():
  key1=rec["team1"]+" Vs "+rec["team2"]
  key2=rec["team2"]+" Vs "+rec["team1"]

  match_id_dict[key1]=rec["match_id"]
  match_id_dict[key2]=rec["match_id"]
match_id_dict
```

```python
# Adding the match_id column in the df_batting dataframe
df_batting["match_id"]=df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
# Converting 4s and 6s column datatype to numeric datatype
df_batting["4s"]=pd.to_numeric(df_batting["4s"])
df_batting["6s"]=pd.to_numeric(df_batting["6s"])
```

```python
# Adding boundary_runs column based on 4s and 6s columns
df_batting["boundary_runs"]= df_batting["4s"]*4+df_batting["6s"]*6
df_batting.head()
```

```python
# Exporting the Dataframe as a csv file
df_batting.to_csv("Batting_summary_csv.csv",index=False)
```

**Bowling Summary**

```python
# Creating a pandas dataframe from the bowling summary json file
with open("/content/t20_wc_bowling_summary.json") as f:
  data=json.load(f)
  all_records=[]
  for rec in data:
    all_records.extend(rec["bowlingSummary"])
df_bowling=pd.DataFrame(all_records)
df_bowling.head()
```

```python
# Adding the match_id column in the df_bowling dataframe
df_bowling["match_id"]=df_bowling["match"].map(match_id_dict)
df_bowling.head()
```

```python
# Removing the unnecessary strings from bowlerName column
df_bowling["bowlerName"]=df_bowling["bowlerName"].apply(lambda x : x.replace("(c)",""))
df_bowling["bowlerName"]=df_bowling["bowlerName"].apply(lambda x : x.replace("",""))

df_bowling.head(10)
```

```python
# Renaming bowlingTeam column to team
df_bowling.rename(columns={"bowlingTeam":"team"},inplace=True)
df_bowling.head()
```

```python
df_bowling.info()
```

```python
# Creating a column based on how many balls the bowler has bowled
df_bowling[['over1', 'over2']] = df_bowling["overs"].str.split(".", expand = True)
df_bowling["over2"].fillna(0,inplace=True)
df_bowling["over1"]=pd.to_numeric(df_bowling["over1"])
df_bowling["over2"]=pd.to_numeric(df_bowling["over2"])

df_bowling["balls"]=(df_bowling["over1"])*6+df_bowling["over2"]
df_bowling.head(25)
```

```python
# Dropping the over1 and over2 columns 
df_bowling.drop(columns={"over1","over2"},inplace=True)
df_bowling.head()
```

```python
# Exporting the Dataframe as a csv file
df_bowling.to_csv("Bowling_summary_csv.csv",index=False)
```

**Player Info**

```python
# Creating a pandas dataframe from the player info json file
with open("/content/t20_wc_player_info.json") as f:
  data = json.load(f)
df_player_info = pd.DataFrame(data)
df_player_info.head(10)
```

```python
# Removing the unnecessary strings from name column
df_player_info["name"]=df_player_info["name"].apply(lambda x : x.replace("(c)",""))
df_player_info["name"]=df_player_info["name"].apply(lambda x : x.replace("",""))

df_player_info.head(10)
```

```python
# Checking whether the table has duplicate records
duplicate = df_player_info[df_player_info.duplicated()]
print(duplicate)
# 6 duplicate values have been found
```

```python
# Removing the duplicate rows
df_player_info.drop_duplicates(keep="last",inplace=True)
df_player_info.head()
```

```python
# Exporting the Dataframe as a csv file
df_player_info.to_csv("Player_info_csv.csv",index=False)
```
